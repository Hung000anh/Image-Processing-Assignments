```python
# Giả sử ảnh có kích thước height x width x 3
height, width, _ = image.shape

# Xác định vị trí trung tâm
center_y = height // 2
center_x = width // 2

# Xác định các chỉ số để lấy vùng ảnh 5x5 từ trung tâm
half_size = 100  # 5x5 => half_size = (5-1)//2
corner_image = image[center_y-half_size:center_y+half_size+1, center_x-half_size:center_x+half_size+1, :]

# Hiển thị vùng ảnh đã cắt
plt.imshow(corner_image[:,:,::-1])
plt.show()
```

```python
# Giả sử img là ảnh gốc và corner_image đã được cắt ra từ giữa ảnh

# Sao chép ảnh gốc
full_image_copy = image.copy()

# Tính toán tọa độ của phần giữa ảnh (vùng đã cắt)
height, width, _ = image.shape
center_y = height // 2
center_x = width // 2
half_size = corner_image.shape[0] // 2

# Thay thế phần giữa bằng màu đỏ (255, 0, 0)
full_image_copy[center_y-half_size:center_y+half_size+1, center_x-half_size:center_x+half_size+1, :] = [255, 0, 0]

# Hiển thị ảnh đã thay đổi
plt.imshow(full_image_copy)
plt.show()
```
