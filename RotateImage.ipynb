```python
#Thêm ảnh 1
iname1="/content/drive/MyDrive/Colab Notebooks/1.png"
image1=cv2.imread(iname1)
plt.imshow(image1)
```

```python
#Thêm ảnh 2
iname2="/content/drive/MyDrive/Colab Notebooks/2.png"
image2=cv2.imread(iname2)
plt.imshow(image2)
```

```python
#xoay anh 1
# Chuyển đổi từ BGR (OpenCV format) sang RGB (matplotlib format)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
# Các góc xoay
angles = [0, 180]

# Tạo biểu đồ với số lượng ô bằng số lượng góc xoay
fig, axs = plt.subplots(1, len(angles), figsize=(15, 5))

# Dùng vòng lặp for để xoay và hiển thị các hình ảnh
for i, angle in enumerate(angles):
    # Lấy kích thước ảnh
    (h, w) = image2.shape[:2]
    center = (w // 2, h // 2)

    # Tạo ma trận xoay
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Thực hiện xoay ảnh
    rotated_image = cv2.warpAffine(image1, M, (w, h))

    # Hiển thị ảnh đã xoay
    axs[i].imshow(rotated_image)
    axs[i].axis('off')

# Hiển thị biểu đồ
plt.show()
```
