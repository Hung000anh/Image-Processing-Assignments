Thư viện sử dụng

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

Đọc ảnh

```python
image = cv2.imread('../XuLyAnhP1/Img/3500-2-.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

Chuyển ảnh thành ảnh trắng đen

```python
#Đặt ngưỡng màu
threshold = 127
thresholded_image = np.where(gray_image > threshold, 255, 0)
```

Hiển thị

```python
plt.imshow(image, cmap='gray')
plt.title('Ảnh gốc')
```

```python
# Ảnh ngưỡng hóa
plt.imshow(thresholded_image, cmap='gray')
plt.title('Ảnh ngưỡng hóa')
```
