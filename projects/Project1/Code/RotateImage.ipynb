```python
#Thêm các thư viện
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Đọc ảnh
image2=cv2.imread('../XuLyAnhP1/Img/anh2.png')
```

```python
#Xoay ảnh
# Chuyển đổi từ BGR (OpenCV format) sang RGB (matplotlib format)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
# Các góc xoay
angles = [0, 180]

# Tạo biểu đồ với số lượng ô bằng số lượng góc xoay
fig, axs = plt.subplots(1, len(angles), figsize=(15, 5))

# Dùng vòng lặp for để xoay và hiển thị các hình ảnh
for i, angle in enumerate(angles):
    # Lấy kích thước ảnh
    (h, w) = image2.shape[:2]
    center = (w // 2, h // 2)

    # Tạo ma trận xoay
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Thực hiện xoay ảnh
    rotated_image = cv2.warpAffine(image2, M, (w, h))

    # Hiển thị ảnh đã xoay
    axs[i].imshow(rotated_image)
    axs[i].axis('off')
```
