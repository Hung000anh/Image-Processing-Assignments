```python
#Thêm các thư viện
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Hàm tạo gradient
def create_and_show_gradient_image(height, width):
    # Tạo một mảng numpy với kích thước chứa giá trị từ 0 đến 255
    gradient = np.linspace(0, 255, width, dtype=np.uint8)

    # Sao chép gradient để tạo ra hình ảnh có nhiều hàng
    image = np.tile(gradient, (height, 1))

    # Hiển thị hình ảnh với Matplotlib
    plt.imshow(image, cmap='gray')
    plt.title("Gradient")
    plt.axis('off') 

# Đặt giá trị chiều cao và chiều rộng của ảnh
height = 300
width = 300

# Gọi hàm để tạo và hiển thị ảnh gradient
create_and_show_gradient_image(height, width)
```
