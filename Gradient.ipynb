```python
#Thêm các thư viện
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Kích thước của hình ảnh (chiều cao, chiều rộng)
height, width = 300, 300

# Tạo một mảng numpy với kích thước chứa giá trị từ 0 đến 255
gradient = np.linspace(0, 255, width, dtype=np.uint8)

# Sao chép gradient để tạo ra hình ảnh có nhiều hàng
image = np.tile(gradient, (height, 1))

# Hiển thị hình ảnh với Matplotlib
plt.imshow(image, cmap='gray')
plt.title("Black to White Gradient")
plt.axis('off')
plt.show()
```

```python
# Kích thước của hình ảnh (chiều cao, chiều rộng)
height, width = 300, 300

# Tạo một mảng numpy với kích thước (height, width) chứa giá trị từ 0 đến 255
gradient = np.linspace(255, 0, height, dtype=np.uint8)

# Sao chép gradient để tạo ra hình ảnh có nhiều cột
image = np.tile(gradient, (width, 1)).T

# Hiển thị hình ảnh với Matplotlib
plt.imshow(image, cmap='gray')
plt.title("Black to White Gradient")
plt.axis('off')
plt.show()
```
