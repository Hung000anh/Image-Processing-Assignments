Lab Task 2:
a) Now mirror the image that you have read at center i.e. the lower half of the image should be the copy
of the upper half. (HINT: You can use nested loops). Write the image to the disk using the right
command.

```python
h, w, _ = image.shape
mirror_img = image.copy()
```

```python
for row in range (int(h/2)): #row chạy được 1 nữa hình
    for col in range (w):    #col chạy hết cột của hình
        mirror_img[h-row -1 , col] = mirror_img[row,col]
```

```python
cv2.imshow("Ảnh gương bán phần", mirror_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
