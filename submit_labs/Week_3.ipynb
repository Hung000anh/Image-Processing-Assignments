1: Apply following transformation techniques on image provided and observe the output image of each transformation.

```python
#Thư viện sử dụng

import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
image = cv2.imread('C:/Users/Student/Downloads/ab67616d0000b2730f40548b1f2bb83b853964cb.jfif', 0)
plt.imshow(image, cmap = "gray")
plt.show()
```

I. Negative transformation

```python
nega_image = (255 - 1) - np.log(image.astype(float) + 1)
# Specify the data type so that
# float value will be converted to int
nega_image = np.array(nega_image, dtype = np.uint8)
plt.imshow(nega_image, cmap = "gray")
plt.show()
```

II. Log transformation

```python
c = 100 / np.log(1 + np.max(image))    
log_image = c * (np.log(image.astype(float) + 1))

# Specify the data type so that
# float value will be converted to int
log_image = np.array(log_image, dtype = np.uint8)
plt.imshow(log_image, cmap = "gray")
plt.show()
```

2: Apply the following transformation on an image.

a) Less than mean => 0
   Greater than mean => 255

```python
mean_img = np.mean(image.astype(float) + 1)
transformed_image = np.where(image < mean_img, 0, 255)
plt.imshow(transformed_image, cmap = "gray")
plt.show()
```

b) Less than mean => 255
Greater than mean => 0

```python
mean_img = np.mean(image.astype(float) + 1)

transformed_image = np.where(image < mean_img, 255, 0)
plt.imshow(transformed_image, cmap = "gray")
plt.show()
```

c) ±20 mean => 0
Otherwise => 255 

```python
mean_img = np.mean(image.astype(float) + 1)
transformed_image = np.where((image >= mean_img - 20) & (image <= mean_img + 20), 255, 0)
plt.imshow(transformed_image, cmap = "gray")
plt.show()
```
